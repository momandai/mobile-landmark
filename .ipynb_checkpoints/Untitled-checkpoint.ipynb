```python
import sys
sys.path.insert(0, '/home/whale/workspace/caffe/python')
import caffe
from skimage import io
import numpy as np
```

```python
prototxt = '/home/whale/workspace/caffe/examples/proto'
model = '/home/whale/workspace/caffe/examples/model'
caffe.set_device(0)
caffe.set_mode_cpu()
net = caffe.Net(prototxt, model, caffe.TEST)
transformer = caffe.io.Transformer({'data': net.blobs['blob1'].data.shape})  #设定图片的shape格式(1,3,28,28) 
transformer.set_transpose('data', (2,0,1))    #改变维度的顺序，由原始图片(28,28,3)变为(3,28,28) 
#transformer.set_mean('data', np.asarray([0.5,0.5,0.5]))    #减去均值，前面训练模型时没有减均值，这儿就不用 
transformer.set_raw_scale('data', 1)    # 缩放到【0，255】之间 
transformer.set_channel_swap('data', (2,1,0))   #交换通道，将图片由RGB变为BGR 
       
im=caffe.io.load_image('/home/whale/test.jpg')                   #加载图片 
net.blobs['blob1'].data[...] = transformer.preprocess('data',im) - np.asarray([0.5])     #执行上面设置的图片预处理操作，并将图片载入到blob中 
print transformer.preprocess('data', im) 
    #执行测试 
out = net.forward() 
```
